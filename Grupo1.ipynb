##Ejercicio N°1:

Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%. (Diversas opciones de solución: libre)

```python
import pandas as pd

inicio = int(input('Introduce el día inicial: '))
fin = inicio + 4
ventas = {}
for i in range(inicio, fin+1):
    ventas[i] = float(input('Introduce las ventas del dia ' + str(i) +': '))
ventas = pd.Series(ventas)
print('Ventas: \n',ventas, '\n')
print('El descuento aplicado a las ventas es: \n',ventas*0.05, '\n')
print('El precio de Venta con Descuento es: \n',ventas*0.95, '\n')
```

##Ejercicio N°2:

Escribir un programa que reciba un diccionario con los ingresos
mensuales de los clientes de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.
El resultado final debe ser una serie con los estadísticos.

```python
import pandas as pd

def sueldos_mensuales(sueldos):
    sueldos = pd.Series(sueldos)
    estadisticos = pd.Series([sueldos.min(), sueldos.max(), sueldos.mean(), ], index=['Ingreso Minimo', 'Ingreso Máximo', 'Media'])
    return estadisticos

sueldos = {'Juan':10000, 'María':20000, 'Pedro':15000, 'Carmen': 24000, 'Luis': 8000}
print(sueldos)
print(sueldos_mensuales(sueldos))
```
