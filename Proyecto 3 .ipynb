```python
# Moda
from collections import Counter

def moda(x):
  c = Counter(x)
  return c.most_common(1)
```

```python
# Centro de amplitud:
def centro_amplitud(x):
  return (max(x) + min(x))/2  # Asignación múltiple
```

```python
df[col_num].apply(centro_amplitud, axis=0).round(2) 
```

```python
# Media geométrica
import numpy as np  # K = platicurtica

def media_geometrica(iterable):
  ''' Calcula la media geométrica'''
  a = np.log(iterable)
  return "Media Geométrica", np.exp(a.sum()/len(a)).round(2) 
```

```python
df[col_num].apply(media_geometrica, axis=0)
```

```python
# Media armónica
def media_armonica(x):
  return "Media Armónica", round(stats.harmonic_mean(x),2)
```

```python
df[col_num].apply(media_armonica, axis=0)
```

```python
# Media recortada al 20%
from scipy import stats as nstats

def media_recortada_20(x):
  mr = nstats.trim_mean(x, proportiontocut=0.1).round(2) # 20%
  return "Media recortada al 20%", mr
```

```python
df[col_num].apply(media_recortada_20, axis=0)
```

```python
# Trimedia
def trimedia(x: pd.Series):
    #Obtenemos los "Quartiles"
    Q1 = x.quantile([.25]).values[0]
    Q2 = x.quantile([.50]).values[0]
    Q3 = x.quantile([.75]).values[0]
    t = (Q1 + 2*Q2 + Q3) / 4
    return t
```

```python
df[col_num].apply(trimedia, axis=0).round(2) 
```

```python
# Rango
def rango(x):
  min = np.min(x)
  max = np.max(x)
  r = max - min
  return min, max, r
```

```python
d = df[col_num].apply(rango, axis=0).T 
d.columns = ["Mínimo","Máximo","Rango"]
d
```

```python
def cuartines(x: pd.Series):
    #Obtenemos los "Quartiles"
    Q1 = x.quantile([.25]).values[0]
    Q2 = x.quantile([.50]).values[0]
    Q3 = x.quantile([.75]).values[0]
    min = np.min(x)
    max = np.max(x)
    return min, Q1, Q2, Q3, max
```

```python
#Obtenemos los Quartiles
d = df[col_num].apply(cuartines, axis=0).T 
d.columns = ["Mínimo","P25% - Q1","P50% - Q2", "P75% - Q3", "Máximo"]
d
```

```python
def rango_intercuartil(x: pd.Series):
    #Obtenemos los "Quartiles"
    Q1 = x.quantile([.25]).values[0]
    Q3 = x.quantile([.75]).values[0]

    return Q1, Q3 - Q1, Q3
```

```python
# Curtosis
from scipy.stats import kurtosis

def kurtosis_data(x):
  curtosis =  kurtosis(x, fisher=True)

  if curtosis >0:
    return ('Es Leptocúrtica-->', round(curtosis,2 ))
  elif curtosis ==0: # 0.03
    return  ('Es Mesocúrtica-->', round(curtosis,2 ))
  else:
    return  ('Es Platicúrtica-->', round(curtosis,2 ))
```

Prueba de normalidad
PLANTEAMIENTO DE LAS HIPóTESIS

Ho: Los datos estan normalmente distribuidos
Ha: Los datos no estan normalmente distribuidos

Nivel de significancia = 5% (0.05)

```python
#import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro # n <= 50
from scipy.stats import kstest  # n > 50
from scipy.stats import zscore
```

```python
# Prueba normalidad: Edad

qqplot(df_f['edad'] , line='s')
plt.show()
```

```python
# Prueba de Shapiro-Wilk (n < =50)
stat, p = shapiro(df_f['edad'])

print('Estadisticos=%.3f, p-value=%.3f' % (stat, p))
```

```python
#Cuando n > 50

stat2, p2 = kstest(df_f['edad'], 'norm')

print('Estadisticos=%.3f, p-value=%.3f' % (stat2, p2))
```

```python
#Decisión: 
#Con un nivel de significancia del 5%, se afirma que, los datos de la variables "Edad" no están normalmente distribuidos. 
```
